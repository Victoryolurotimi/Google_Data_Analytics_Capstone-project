### Data Manipulation

#### steps:

* Change  ActivityDate to datatime64 dtype.


* Change format of ActivityDate to yyyy-mm-dd.


* Create a new column "**weekday**" by separating the date into  each day of the week.


* Create a new column "**TotalMinutes**"which is the sum of "**VeryActiveMinutes**", "**FairlyActiveMinutes**", "**LightlyActiveMinutes**" and "**SedentaryMinutes**".


* Create new column TotalHours by converting new column above to number of hours.


* Rearrange and rename columns.

```python
# change ActivityDate to datatime64 dtype.

daily_activity["ActivityDate"] = pd.to_datetime(daily_activity["ActivityDate"], format="%m/%d/%Y")

# confirm the change done 

daily_activity.info()

# print the activityDate to be sure

daily_activity["ActivityDate"].head(10)
```

```python
# create new column "weekday" to represent each day of the week 
daily_activity["weekday"] = daily_activity["ActivityDate"].dt.day_name()

# print 1st 5 rows to confirm
daily_activity["weekday"].head(7)
```

* New columns would be added such as: '**weekday**' , '****TotalExerciseMinutes**' and '**TotalExerciseHours**' 
    
    
* To account for the new columns (**new_cols**) to be added to the old list, there would  be a new list called (**everyday_activity**)

```python
#r create new list of rearranged columns
new_cols = ['Id', 'ActivityDate', 'weekday', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalExerciseMinutes', 'TotalExerciseHours', 'Calories']

# accounting for the new columns based on "new_cols"
everyday_activity = daily_activity.reindex(columns=new_cols)

# print 1st 5 rows to confirm
everyday_activity.head(10)
```

To be able to get values for the newly added columns, there has to be summation function to add for each column,for example; 

* TotalExerciseMinutes = '**VeryActiveMinutes**' + '**FairlyActiveMinutes**' + '**LightlyActiveMinutes**' + '**SedentaryMinutes**'.


* **TotalExeciseHours** =  '**TotalExerciseMinutes' / 60**

```python
# To get values for "totalExerciseMinutes" containing sum of totalExerciseMinutes.

everyday_activity["TotalExerciseMinutes"] = everyday_activity["VeryActiveMinutes"] + everyday_activity["FairlyActiveMinutes"] + everyday_activity["LightlyActiveMinutes"] + everyday_activity["SedentaryMinutes"]
everyday_activity["TotalExerciseMinutes"].head(10)
```

```python
# To get values for "totalExerciseHours" containing sum of totalExerciseHours in the nearest integer.

everyday_activity["TotalExerciseHours"] = round(everyday_activity['TotalExerciseMinutes'] / 60)
everyday_activity["TotalExerciseHours"].head(10)
```

# PHASE 5: SHARE
 
#### Data Visualization

This is the phase where insights are drawn and shown in charts and visulized. 

```python
# import matplotlib package
import matplotlib.pyplot as plt

# plotting histogram
plt.figure(figsize=(8,6)) 
plt.hist(everyday_activity.weekday, bins = 7, width = 0.6, color = "blue")

# adding labels
plt.xlabel("weekday")
plt.ylabel("Frequency")
plt.title("times users had their smart watches on")
plt.grid(True)
plt.show()
```

* We can deduce from the above histogram visualization shows a decline in the number of times the users wear their smart watches in a day. 
* This makes Tuesday the day people use their smart watches the most.
* And during the weekend, people tend to not wear their smart watches probably due to less activity during the weekend.

```python
import matplotlib.pyplot as plt

# summing up VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes
VeryActiveMinutes = everyday_activity['VeryActiveMinutes'].sum()
FairlyActiveMinutes = everyday_activity['FairlyActiveMinutes'].sum()
LightlyActiveMinutes = everyday_activity['LightlyActiveMinutes'].sum()
SedentaryMinutes = everyday_activity['SedentaryMinutes'].sum()

# setting piechart
slices = [VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes]
labels= ["VeryActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes", "SedentaryMinutes"]
colours = ["pink", "lightblue", "yellow", "purple"]
explode = [0, 0, 0, 0]
plt.style.use("default")
plt.pie(slices, labels = labels, colors = colours, explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```

* As seen from the pie chart above, sedentary minutes takes the largest chunk of the circle.This proves that majority of the times the fitBit smart watches were worn, the users weren't using them for active activities like running,jogging or exercises.


* This implies that the smart watches aren't used by the users for enhancing general body fitness. 


```python
# import matplotlib package
import matplotlib.pyplot as plt

# plotting scatter plot
plt.style.use("default")
plt.figure(figsize=(8,6)) 
plt.scatter(everyday_activity.TotalSteps, everyday_activity.Calories, 
            alpha = 0.8, c = everyday_activity.Calories, 
            cmap = "viridis")

median_calories = 2303
median_steps = 7637

plt.colorbar(orientation = "vertical")
plt.axvline(median_steps, color = "Blue", label = "Median steps")
plt.axhline(median_calories, color = "Red", label = "Median calories burned")
plt.xlabel("Steps")
plt.ylabel("Calories burnt")
plt.title("Calories burnt per step")
plt.grid(True)
plt.legend()
plt.show()
```

* It is seen above that the more steps the user took,the more the calories burnt. This is a positive correlation


* However, there are some outliers in the graph,at over 35000 steps, close to 3000 calories were burnt,this presumed to be a miscalculation or an error during data entry.
